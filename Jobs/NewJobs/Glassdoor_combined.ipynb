```python
import pandas as pd
from IPython.core.display import display

jobs = pd.read_csv("glassdoor_jobs.csv")
salaries = pd.read_csv("glassdoor_salaries.csv")

jobs['Company Name'] = jobs['Company Name'].str.lower()
jobs['Company Name'] = jobs['Company Name'].str.split("\n").str[0]

jobs
```

```python
salaries['Company Name'] = salaries['company'].str.lower()
salaries
```

```python
finaloutput = pd.merge(jobs, salaries, how='outer', on=['Company Name'])
finaloutput
```

```python
finaloutput.to_csv("glassdoor_combined.csv", index = False)
```
